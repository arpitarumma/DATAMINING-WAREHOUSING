```python
# Import necessary libraries
import pandas as pd
!pip install mlxtend
from mlxtend.frequent_patterns import fpgrowth

# Display the first few rows to understand the structure
print(df.head())

# Apply the FP-Growth algorithm
min_support = 2 / len(df)

# Create a one-hot encoded DataFrame using TransactionEncoder
from mlxtend.preprocessing import TransactionEncoder
encoder = TransactionEncoder()
onehot = encoder.fit(df['Items Purchased']).transform(df['Items Purchased'])
onehot_df = pd.DataFrame(onehot, columns=encoder.columns_)
onehot_df = onehot_df.astype(int)

# Use the one-hot encoded DataFrame for fpgrowth
frequent_itemsets = fpgrowth(onehot_df, min_support=min_support, use_colnames=True)

# Display the frequent itemsets
print(frequent_itemsets)
```
